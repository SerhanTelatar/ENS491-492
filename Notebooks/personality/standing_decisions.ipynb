```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
personality = pd.read_excel("../../Datasets/Personality Survey Data.xlsx")
```

```python
standing_decisions = personality.iloc[253:299, :]
standing_decisions = standing_decisions.iloc[:,2:]
standing_decisions.columns = personality.iloc[50, 2:]
standing_decisions.reset_index(drop=True, inplace = True)
```

```python
standing_decisions
```

```python
faculty = pd.get_dummies(standing_decisions['Faculty'])
faculty = faculty.replace({True: 1, False: 0})
faculty.columns = ['faculty_' + col for col in faculty.columns]
standing_decisions = pd.concat([standing_decisions, faculty], axis=1)
standing_decisions.drop(["Faculty"], axis=1,inplace = True)
```

```python
program = pd.get_dummies(standing_decisions['Program'])
program = program.replace({True: 1, False: 0, })
program.columns = ['program_' + col for col in program.columns]
standing_decisions = pd.concat([standing_decisions, program], axis=1)
standing_decisions.drop(["Program"], axis=1,inplace = True)
```

```python
Grad_Undergrad = pd.get_dummies(standing_decisions['Grad/ Undergrad'])
Grad_Undergrad = Grad_Undergrad.replace({True: 1, False: 0})
Grad_Undergrad.columns = ['Grad_Undergrad_' + col for col in Grad_Undergrad.columns]
standing_decisions = pd.concat([standing_decisions, Grad_Undergrad], axis=1)
standing_decisions.drop(["Grad/ Undergrad"], axis=1,inplace = True)
```

```python
standing_decisions["ENS 208"] = standing_decisions["ENS 208"].replace({"YES": 1, np.nan: 0}).astype(int)
standing_decisions["MS 401"] = standing_decisions["MS 401"].replace({"YES": 1, np.nan: 0}).astype(int)
standing_decisions["Other Supply Chain Course?"] = standing_decisions["Other Supply Chain Course?"].replace({"YES": 1, np.nan: 0}).astype(int)
standing_decisions
```
