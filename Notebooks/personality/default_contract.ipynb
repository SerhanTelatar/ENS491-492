```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
personality = pd.read_excel("../../Datasets/Personality Survey Data.xlsx")
```

```python
default_contract = personality.iloc[152:196, :]
default_contract = default_contract.iloc[:,2:]
default_contract.columns = personality.iloc[50, 2:]
default_contract.reset_index(drop=True, inplace = True)
```

```python
default_contract
```

```python
faculty = pd.get_dummies(default_contract['Faculty'])
faculty = faculty.replace({True: 1, False: 0})
faculty.columns = ['faculty_' + col for col in faculty.columns]
default_contract = pd.concat([default_contract, faculty], axis=1)
default_contract.drop(["Faculty"], axis=1,inplace = True)
```

```python
program = pd.get_dummies(default_contract['Program'])
program = program.replace({True: 1, False: 0, })
program.columns = ['program_' + col for col in program.columns]
default_contract = pd.concat([default_contract, program], axis=1)
default_contract.drop(["Program"], axis=1,inplace = True)
```

```python
Grad_Undergrad = pd.get_dummies(default_contract['Grad/ Undergrad'])
Grad_Undergrad = Grad_Undergrad.replace({True: 1, False: 0})
Grad_Undergrad.columns = ['Grad_Undergrad_' + col for col in Grad_Undergrad.columns]
default_contract = pd.concat([default_contract, Grad_Undergrad], axis=1)
default_contract.drop(["Grad/ Undergrad"], axis=1,inplace = True)
```

```python
default_contract["ENS 208"] = default_contract["ENS 208"].replace({"YES": 1, np.nan: 0}).astype(int)
default_contract["MS 401"] = default_contract["MS 401"].replace({"YES": 1, np.nan: 0}).astype(int)
default_contract["Other Supply Chain Course?"] = default_contract["Other Supply Chain Course?"].replace({"YES": 1, np.nan: 0}).astype(int)
default_contract
```
