```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
personality = pd.read_excel("../../Datasets/Personality Survey Data.xlsx")
```

```python
standing_contract = personality.iloc[202:246, :]
standing_contract = standing_contract.iloc[:,2:]
standing_contract.columns = personality.iloc[50, 2:]
standing_contract.reset_index(drop=True, inplace = True)
```

```python
faculty = pd.get_dummies(standing_contract['Faculty'])
faculty = faculty.replace({True: 1, False: 0})
faculty.columns = ['faculty_' + col for col in faculty.columns]
standing_contract = pd.concat([standing_contract, faculty], axis=1)
standing_contract.drop(["Faculty"], axis=1,inplace = True)
```

```python
program = pd.get_dummies(standing_contract['Program'])
program = program.replace({True: 1, False: 0, })
program.columns = ['program_' + col for col in program.columns]
standing_contract = pd.concat([standing_contract, program], axis=1)
standing_contract.drop(["Program"], axis=1,inplace = True)
```

```python
Grad_Undergrad = pd.get_dummies(standing_contract['Grad/ Undergrad'])
Grad_Undergrad = Grad_Undergrad.replace({True: 1, False: 0})
Grad_Undergrad.columns = ['Grad_Undergrad_' + col for col in Grad_Undergrad.columns]
standing_contract = pd.concat([standing_contract, Grad_Undergrad], axis=1)
standing_contract.drop(["Grad/ Undergrad"], axis=1,inplace = True)
```

```python
standing_contract["ENS 208"] = standing_contract["ENS 208"].replace({"YES": 1, np.nan: 0}).astype(int)
standing_contract["MS 401"] = standing_contract["MS 401"].replace({"YES": 1, np.nan: 0}).astype(int)
standing_contract["Other Supply Chain Course?"] = standing_contract["Other Supply Chain Course?"].replace({"YES": 1, np.nan: 0}).astype(int)
standing_contract
```
