```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
personality = pd.read_excel("../../Datasets/Personality Survey Data.xlsx")
```

```python
revenue_sharing_contract = personality.iloc[100:144, :]
revenue_sharing_contract = revenue_sharing_contract.iloc[:,2:]
revenue_sharing_contract.columns = personality.iloc[50, 2:]
revenue_sharing_contract.reset_index(drop=True, inplace = True)
```

```python
faculty = pd.get_dummies(revenue_sharing_contract['Faculty'])
faculty = faculty.replace({True: 1, False: 0})
faculty.columns = ['faculty_' + col for col in faculty.columns]
revenue_sharing_contract = pd.concat([revenue_sharing_contract, faculty], axis=1)
revenue_sharing_contract.drop(["Faculty"], axis=1,inplace = True)
```

```python
program = pd.get_dummies(revenue_sharing_contract['Program'])
program = program.replace({True: 1, False: 0, })
program.columns = ['program_' + col for col in program.columns]
revenue_sharing_contract = pd.concat([revenue_sharing_contract, program], axis=1)
revenue_sharing_contract.drop(["Program"], axis=1,inplace = True)
```

```python
Grad_Undergrad = pd.get_dummies(revenue_sharing_contract['Grad/ Undergrad'])
Grad_Undergrad = Grad_Undergrad.replace({True: 1, False: 0})
Grad_Undergrad.columns = ['Grad_Undergrad_' + col for col in Grad_Undergrad.columns]
revenue_sharing_contract = pd.concat([revenue_sharing_contract, Grad_Undergrad], axis=1)
revenue_sharing_contract.drop(["Grad/ Undergrad"], axis=1,inplace = True)
```

```python
revenue_sharing_contract["ENS 208"] = revenue_sharing_contract["ENS 208"].replace({"YES": 1, np.nan: 0}).astype(int)
revenue_sharing_contract["MS 401"] = revenue_sharing_contract["MS 401"].replace({"YES": 1, np.nan: 0}).astype(int)
revenue_sharing_contract["Other Supply Chain Course?"] = revenue_sharing_contract["Other Supply Chain Course?"].replace({"YES": 1, np.nan: 0}).astype(int)
revenue_sharing_contract
```
